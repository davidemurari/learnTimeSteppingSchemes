```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib

matplotlib.rc('font', size=15)

plt.rcParams.update({
    'text.usetex': True,
    'font.family': 'serif',
    'text.latex.preamble': r'\usepackage[T1]{fontenc}',
})
```

```python
T = 2
N = 4
x = np.linspace(0,T,N+1)
y = np.zeros_like(x)  # Assuming all y values are 0 as it appears in the image.

fig = plt.figure(figsize=(10,5),dpi=600)

plt.plot(x, y, 'k-o')

# Annotating each point with its x value
for i, value in enumerate(x):
    plt.annotate(f'{value}', (x[i], y[i]), textcoords="offset points", xytext=(0,15), ha='center')

xk = np.linspace(x[2],x[3],N+1)
yk = np.zeros_like(xk)
plt.plot(xk,yk,'ro')

# Annotating each point with its x value
for i, value in enumerate(xk[1:-1]):
    sign = (-1)**(i+1)
    correction = 0 if sign==1 else -10
    plt.annotate(f'{value}', (xk[i+1], yk[i+1]), textcoords="offset points", xytext=(0,sign*15+correction), ha='center')

plt.axis('off')
plt.savefig("collocationPoints.pdf",bbox_inches='tight')
```
